### Visualization Section
### Plot 1: Distribution of House Prices (Histogram)

```python
plt.figure(figsize=(10,6))
sns.histplot(df['SalePrice'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of House Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

### Insight:
Most houses are priced between 100,000 and 200,000.  
The distribution is right-skewed, meaning there are a few very expensive houses.


### Plot 2: Living Area vs. Sale Price (Scatter Plot)

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, color='orange')
plt.title('Living Area vs. Sale Price')
plt.xlabel('Ground Living Area (sq ft)')
plt.ylabel('Sale Price')
plt.show()
```

### Insight:
Larger houses generally sell for higher prices, showing a clear positive correlation between living area and price.

### Plot 3: Overall Quality vs. Sale Price (Box Plot)

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='OverallQual', y='SalePrice', data=df, palette='viridis')
plt.title('Sale Price by Overall Quality')
plt.xlabel('Overall Quality')
plt.ylabel('Sale Price')
plt.show()
```

### Insight:
Houses with higher overall quality tend to have significantly higher sale price

### Plot 4: Correlation Heatmap

```python
plt.figure(figsize=(12, 8))

# Select only numeric columns
corr = df.select_dtypes(include=['number']).corr()

sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

### Insight:
The most correlated features with SalePrice include OverallQual, GrLivArea, and GarageCars. These will be useful in model building later.

### Plot 5: Average Price by Garage Capacity (Bar Plot)

```python
plt.figure(figsize=(8,6))
sns.barplot(x='GarageCars', y='SalePrice', data=df, palette='pastel')
plt.title('Average Sale Price by Garage Capacity')
plt.xlabel('Number of Garage Cars')
plt.ylabel('Average Sale Price')
plt.show()
```

### Insight:
Houses with more garage spaces generally have higher sale prices, showing that garage capacity adds property value.

### Summary of Insights


- House prices are right-skewed, with most homes priced below 200,000.  
- Larger houses and higher quality ratings strongly increase sale prices.  
- Features like `OverallQual`, `GrLivArea`, and `GarageCars` have the strongest positive correlation with price.  
- These insights will guide feature selection for model building in the coming weeks.
